```python
import numpy as np
import matplotlib.pyplot as plt

def lorenz(x, y, z, sigma=10, rho=28, beta=8/3):
    deriv_x = sigma*(y - x)
    deriv_y = rho*x - y - x*z
    deriv_z = x*y - beta*z
    return deriv_x, deriv_y, deriv_z

dt = 0.01
nb_pas = 6000
x = np.empty(nb_pas + 1)
y = np.empty(nb_pas + 1)
z = np.empty(nb_pas + 1)

sigma = 10
beta = 8/3
rhos = [10, 20, 27, 30]  # Valeurs de rho  tester

for rho in rhos:
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    
    # Conditions initiales pour les deux points de dpart
    x[0], y[0], z[0] = (0, 1, 0)  # Point de dpart bleu
    for i in range(nb_pas):
        deriv_x, deriv_y, deriv_z = lorenz(x[i], y[i], z[i], sigma, rho, beta)
        x[i + 1] = x[i] + (deriv_x * dt)
        y[i + 1] = y[i] + (deriv_y * dt)
        z[i + 1] = z[i] + (deriv_z * dt)
    axs[0].plot(x, z, color='darkorchid')
    axs[1].plot(np.arange(len(x)), x, color='darkorchid')

    x[0], y[0], z[0] = (0, 1.01, 0)  # Point de dpart rouge
    for i in range(nb_pas):
        deriv_x, deriv_y, deriv_z = lorenz(x[i], y[i], z[i], sigma, rho, beta)
        x[i + 1] = x[i] + (deriv_x * dt)
        y[i + 1] = y[i] + (deriv_y * dt)
        z[i + 1] = z[i] + (deriv_z * dt)
    axs[0].plot(x, z, color='mediumseagreen')
    axs[1].plot(np.arange(len(x)), x, color='mediumseagreen')

    axs[0].set_xlabel('X')
    axs[0].set_ylabel('Z')
    axs[0].set_title('Plan XZ, rho = {}'.format(rho))

    axs[1].set_xlabel('Temps')
    axs[1].set_ylabel('X')
    axs[1].set_title('Temps en fonction X, rho = {}'.format(rho))

    plt.tight_layout()
    plt.show()
```
