```python
import numpy as np
import matplotlib.pyplot as plt

def lorenz(x, y, z, sigma=10, rho=28, beta=2.667):
    deriv_x = sigma*(y - x)
    deriv_y = rho*x - y - x*z
    deriv_z = x*y - beta*z
    return deriv_x, deriv_y, deriv_z
dt = 0.01
nb_pas = 1500
x = np.empty(nb_pas + 1)
y = np.empty(nb_pas + 1)
z = np.empty(nb_pas + 1)

x[0], y[0], z[0] = (1.0, 3.0, 5)
for i in range(nb_pas):
    deriv_x, deriv_y, deriv_z = lorenz(x[i], y[i], z[i])
    x[i + 1] = x[i] + (deriv_x * dt)
    y[i + 1] = y[i] + (deriv_y * dt)
    z[i + 1] = z[i] + (deriv_z * dt)
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(x, y, z,s = 2)
ax.plot(x, y, z, color = 'r')
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

```python

fig, ax = plt.subplots(1, 3, figsize=(12, 3))

# plot x-y phase plane
ax[0].plot(x, y, color='b', alpha=1, linewidth=0.6)
ax[0].set_title('x-y phase plane')

# plot x-z phase plane
ax[1].plot(x, z, color='g', alpha=1, linewidth=0.6)
ax[1].set_title('x-z phase plane')

# plot y-z phase plane
ax[2].plot(y, z, color='c', alpha=1, linewidth=0.6)
ax[2].set_title('y-z phase plane')

plt.show()
```
