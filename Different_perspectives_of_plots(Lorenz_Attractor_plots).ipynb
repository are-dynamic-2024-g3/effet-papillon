```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

sigma = 10
rho = 28
beta = 2.667

def lorenz(etat, t, sigma, rho, beta):
    x, y, z = etat
    deriv_x = sigma*(y - x)
    deriv_y = rho*x - y - x*z
    deriv_z = x*y - beta*z
    return [deriv_x, deriv_y, deriv_z]

dt = 0.01
t = np.arange(0, dt*1550, dt)

etat_initial = [0.0, 2.0, 6.0]

solution = odeint(lorenz, etat_initial, t, args=(sigma, rho, beta))

x = solution[:, 0]
y = solution[:, 1]
z = solution[:, 2]
# Deuxième simulation
dt2 = 0.01
t2 = np.arange(0, dt2*1550, dt2)
etat_initial2 = [0.1, 2.1, 6.1]
solution2 = odeint(lorenz, etat_initial2, t2, args=(sigma, rho, beta))
x2 = solution2[:, 0]
y2 = solution2[:, 1]
z2 = solution2[:, 2]

# Création de la figure et des sous-graphiques
fig, axes = plt.subplots(1, 2, figsize=(12, 5), subplot_kw={'projection':'3d'})

# Graphique 1
axes[0].scatter(x, y, z, s=2)
axes[0].plot(x, y, z, color='r', alpha=0.5)
axes[0].set_xlabel("X Axis")
axes[0].set_ylabel("Y Axis")
axes[0].set_zlabel("Z Axis")
axes[0].set_title("Lorenz Attractor 1")

# Graphique 2
axes[1].scatter(x2, y2, z2, s=2)
axes[1].plot(x2, y2, z2, color='r', alpha=0.5)
axes[1].set_xlabel("X Axis")
axes[1].set_ylabel("Y Axis")
axes[1].set_zlabel("Z Axis")
axes[1].set_title("Lorenz Attractor 2")

plt.tight_layout()
plt.show()



fig2, ax = plt.subplots(1, 3, figsize=(10, 3))

ax[0].plot(t,x, color='r', alpha=1, linewidth=0.6)
ax[0].set_xlabel("TIME Axis")
ax[0].set_ylabel("X Axis")

ax[1].plot(t,y, color='b', alpha=1, linewidth=0.6)
ax[1].set_xlabel("TIME Axis")
ax[1].set_ylabel("Y Axis")


ax[2].plot(t,z, color='g', alpha=1, linewidth=0.6)
ax[2].set_xlabel("TIME Axis")
ax[2].set_ylabel("Z Axis")


fig3, ax = plt.subplots(1, 3, figsize=(10, 3))

ax[0].plot(t,x2, color='r', alpha=1, linewidth=0.6)
ax[0].set_xlabel("TIME Axis")
ax[0].set_ylabel("X2 Axis")

ax[1].plot(t,y2, color='b', alpha=1, linewidth=0.6)
ax[1].set_xlabel("TIME Axis")
ax[1].set_ylabel("Y2 Axis")


ax[2].plot(t,z2, color='g', alpha=1, linewidth=0.6)
ax[2].set_xlabel("TIME Axis")
ax[2].set_ylabel("Z2 Axis")




plt.show()
```
